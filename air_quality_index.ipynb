```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
plt.plot(air_quality_data['state'],air_quality_data['pollutant_avg'])
plt.xlabel("state")
plt.ylabel("pollutant_avg")
plt.xticks(rotation=90)
plt.title('state vs pollutant_avg')
plt.figure(figsize=(12,6))
```

```python
#converting categorical values into numerical for ml
```

```python
air_quality_data['pollutant_id']=air_quality_data['pollutant_id'].astype('category').cat.codes
```

```python
air_quality_data['pollutant_id'].head(100)
```

```python
x=air_quality_data[['pollutant_id']]
y=air_quality_data['pollutant_avg']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
model.predict(x_test)
```

```python
predictions=model.predict(x_test)
results = pd.DataFrame({
    'pollutant_id': x_test.squeeze(),
    'perdicted pollutant_avg' : predictions
})
results.head()
```

```python
res=pd.DataFrame({
    'pollutant_id':x_test.squeeze(),
    'Actual pollutant_avg': y_test,
    'predicted pollutant_avg': predictions
})
res.head()
```

```python
#evaluating accuracy
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions)
print("Mean Squared Error:",mse)
```

```python
plt.scatter(y_test,predictions,color='orange')
plt.plot([y_test.min(),y_test.max()],[predictions.min(),predictions.max()],color='red', linestyle='--')
plt.xlabel('actual values')
plt.ylabel('predicted values')
plt.title("actual vs predicted values")
```

```python
#classification of states into low ,moderate and high air quality based on the pollutant avg
```

```python
def classify_quality(value):
  if value < 50:
    return 'low'
  elif value < 100:
    return 'moderate'
  else:
    return 'High'
air_quality_data['air_quality']=air_quality_data['pollutant_avg'].apply(classify_quality)
air_quality_data[['state','city','pollutant_avg','air_quality']].head(1000)
```
